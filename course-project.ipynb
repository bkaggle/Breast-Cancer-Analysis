# Q3)Survival chance by age?

```python
#The FacetGrid class is useful when you want to visualize the distribution of a variable or the relationship between multiple variables separately within subsets of your dataset.

a=sns.FacetGrid(cancer_df,hue='Patient_Status',height=5)
a.map(sns.distplot,'Age')
a.add_legend()
```

```python
#Age groups from 30–40 have equal chances of survival, 
#40–60 have high chances of survival & 
#age groups of 60–75 have less chances of survival.
```
